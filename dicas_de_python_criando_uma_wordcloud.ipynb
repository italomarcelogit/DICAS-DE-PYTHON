# Criando uma wordcloud

```python
# Carregando letras
import pandas as pd
dataset = 'https://github.com/italomarcelogit/scripts-diversos/raw/master/dataset-lyrics/dataset-lyrics-musics-mini.csv'
letras = pd.read_csv(dataset, nrows=20, usecols=['letra'])
```

```python
# criando uma lista de palavras encontradas nas 20 letras lidas acima
lista_de_palavras = []
for x in range(0, letras.shape[0]):
  linha = str(letras['letra'].loc[x]).replace('.','').replace(',','').split(' ')
  lista_de_palavras += linha
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
stopwords = set(STOPWORDS)
import numpy as np
from PIL import Image
```

```python
# função para gerar a wordcloud
def myWordCloud(data, title, width = 900, height = 500,):
  
  wordcloud = WordCloud(width = width, height = height, 
                        background_color ='white', 
                        colormap='Set2', collocations=False,
                        stopwords = stopwords, 
                        min_font_size = 10).generate(' '.join(data)) 
  
  plt.figure(figsize = (6, 3), facecolor = None) 
  plt.imshow(wordcloud) 
  plt.axis("off") 
  plt.tight_layout(pad = 0) 
  plt.title(title)
  plt.show()
```

```python
  # Exibindo a wordcloud
  myWordCloud(lista_de_palavras, 'Minha WordCloud')
```
